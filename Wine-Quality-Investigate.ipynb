## 合并数据处理

### 创建颜色列

# 实际问题解决
## 问题 1：某种类型的葡萄酒（红葡萄酒或白葡萄酒）是否代表更高的品质？
    思路:将红葡萄酒的平均质量与白葡萄酒的平均质量等级进行比较。

```python
# 先按颜色分组，然后找到每组的平均质量等级
level = wine_quality.groupby('color').mean()['quality']
level
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
level.plot(kind='bar',title='Average Wine Quality By Color',color=('g','b'),alpha=0.6)
plt.xlabel('Color',fontsize=17)
plt.ylabel('Quality',fontsize=15)
```

初步结论: 从图形可以看出,白葡萄酒的平均质量等级是高于红葡萄酒的

追问: 从整体的等级评分来看,都是白葡萄质量高于红葡萄酒吗?

```python
# 验证以上的追问问题
# 获取每种酒的每个等级的统计数据
count_ph = wine_quality.groupby(['color','quality']).count()['pH']
count_ph
```

思路 :  因为红葡萄酒和白葡萄酒数量不一致,不能直接用于比较,所以,需要转换一下,进行比例的比较
    1. 红葡萄酒条柱比例 = 每个质量等级的数量 / 红葡萄酒样本的总数
 
    2.白葡萄酒条柱比例 = 每个质量等级的数量 / 白葡萄酒样本的总数

```python
# 计算各葡萄酒的总数
total_color = wine_quality.groupby('color').count()['pH']
```

```python
# 将红葡萄酒等级数量除以红葡萄酒样本总数，获取比例
red_level = count_ph['RED'] / total_color['RED']
white_level = count_ph['WHITE'] / total_color['WHITE']
red_level
white_level
```

    # 提示 白葡萄酒有等级为 9 的数据,但是红葡萄酒缺失,所以,需要加上红葡萄等级 9 的数据:0

```python
# 添加红葡萄酒等级为 9 的数据
red_level['9']=0
red_level
```

# 绘制图表

```python
ind = np.arange(len(red_level))
width = 0.45

red_bar = plt.bar(ind,red_level, width,alpha=0.7,label='Red Wine')
white_bar = plt.bar(ind + width, white_level,width,alpha=0.7,label='White Wine')

plt.ylabel('Proportion')
plt.xlabel('Quality')
plt.title('Proportion by Wine Color and Quality')
locations = ind + width /2
labels =['3','4','5','6','7','8','9']
plt.xticks(locations,labels)

plt.legend()
```

# 分析与结论：
根据可视化图形可以得出：
1.从平均值来看，白葡萄酒的质量稍比红葡萄酒高；

2.后续优化图能得到：在质量为3,4,5 时，红葡萄酒的质量是高于白葡萄酒的；

3.当质量升高到6,7,8,9 时，红葡萄酒的质量明显低于白葡萄酒。

# 问题二 酒精含量高底是否会影响评分的高低？
    # 思路: 1.分别建立酒精含量低和酒精含量高的两组数据
            2.分别对两组数据进行评分高低的分析

```python
# 获取酒精含量的中位数
median = df['alcohol'].median()
median
```

```python
# 分别建立酒精含量低和酒精含量高的两组数据
low = df.query('alcohol < {}'.format(median))
high = df.query('alcohol >= {}'.format(median))
```

```python
# 计算酒精含量高和酒精含量低的评分均值
low_quality_data = low['quality'].mean()
high_quality_data = high['quality'].mean()
high_quality_data
```

```python
low_quality_data
```

```python
# 根据以上统计数据绘制图形
locations = [1,2]
heights =[low_quality_data,high_quality_data]
labels = ['Low','High']
plt.bar(locations,heights,tick_label = labels)
# low_bar = plt.bar(low_quality_data,label='Low Quality')
# hight_bar = plt.bar(high_quality_data,label = 'High Quality')
plt.xlabel('Species')
plt.ylabel('Quality')
plt.title('Quality By Alcohol')
```
